```python
import pandas as pd
import numpy as np
Data_Set=pd.read_excel("Bike_Sales_Functions_Lab.xlsx")
Data_Set.head(20)

```

```python
duplicates=Data_Set.duplicated("Sales_Order #")
duplicates
```

```python
Total_duplicates=duplicates.sum()
Total_duplicates
```

```python
# droping duplicates
Drop_Duplicates=Data_Set.drop_duplicates("Sales_Order #")
Drop_Duplicates.head(20)
```

```python
# finding null values in the data set
Null_values=Data_Set.isnull()
```

```python
# Total null values in data_set
Total_null_values=Null_values.sum()
Total_null_values
```

```python
#  handling customer gender column nan values using bfill method
Data_Set["Customer_Gender"]=Data_Set["Customer_Gender"].fillna(method="bfill")
Data_Set.head(20)
```

```python

print(Data_Set.columns.str.strip())
```

```python
print(list(Data_Set.columns))
Data_Set.rename(columns={"Unit_Cost ":"Unit_Cost","Unit_Price ":"Unit_Price",'Cost ':"Cost"},inplace=True)
Data_Set.head().columns
```

```python
# Replacing null value with mean of Unit_cost, Unit_price and Cost
print("Mean of Unit Cost:",Data_Set["Unit_Cost"].mean())
print("Mean of Unit Price:",Data_Set["Unit_Price"].mean())
```

```python
# replacing null with means
Data_Set["Unit_Cost"]=Data_Set["Unit_Cost"].replace(np.nan,1067)
Data_Set["Unit_Price"]=Data_Set["Unit_Price"].replace(np.nan,1973)
```
